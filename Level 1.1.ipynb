```python
Cuisines=df['Cuisines'].value_counts()
Cuisines.head(3)
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x=Cuisines.head(3).index, y=Cuisines.head(3).values, palette='viridis')
plt.title('Top 3 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
```

```python

split = df['Cuisines'].str.split(',').explode()
counts =split.value_counts()
percentages = (counts / len(df)) * 100
print("Percentage of Restaurants Serving Top Cuisine:")
print(percentages)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(10, 6))

# Create bar plot (Ensure 'Cuisines' is a Pandas Series)
sns.barplot(
    x=Cuisines.head(10).index,
    y=Cuisines.head(10).values,
    palette='viridis'
)

# Add title and labels
plt.title('Top 10 Cuisines and Their Percentages')
plt.xlabel('Cuisines')
plt.ylabel('Percentage')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
